## Observation de l'extrait : WorldCupMatches2014

Colonnes principales :

  - Year, Datetime, Stage → contexte du match.
  - Stadium, City → localisation.
  - Home Team Name, Away Team Name → équipes.
  - Home Team Goals, Away Team Goals → scores.
  - Win conditions → prolongations, tirs au but (souvent NaN si victoire classique).
  - Attendance → nombre de spectateurs (parfois manquant).
  - Half-time Home Goals, Half-time Away Goals → score à la mi-temps.
  - Referee, Assistant 1, Assistant 2 → arbitres.
  - RoundID, MatchID → identifiants techniques.
  - Home Team Initials, Away Team Initials → codes FIFA.
