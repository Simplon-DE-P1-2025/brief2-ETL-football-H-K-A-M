```python
import pandas as pd

# Charger les datasets
df1 = pd.read_csv("../data/raw/matches_19302010.csv")
df2 = pd.read_csv("../data/processed/matches_2022.csv")
df3 = pd.read_csv("../data/raw/Fifa_world_cup_matches.csv")
df4 = pd.read_csv("../data/raw/world_cup_data_2022_stadium.csv")
df5 = pd.read_csv("../data/raw/WorldCupMatches2014-WorldCupMatches2014.csv")

datasets = {
    "1930-2010": df1,
    "2022": df2,
    "Fifa Kaggle": df3,
    "Stadiums 2022": df4,
    "WorldCup 2014": df5
}

pd.set_option("display.max_rows", None)      # affiche toutes les lignes
pd.set_option("display.max_columns", None)   # affiche toutes les colonnes
pd.set_option("display.width", None)         # pas de limite de largeur
pd.set_option("display.max_colwidth", None)  # pas de limite sur la taille des cellules


# Boucle d’observation
for name, df in datasets.items():
    print(f"\n=== Dataset {name} ===")
    print("Shape :", df.shape)
    print("Types des colonnes :")
    print(df.dtypes)
    print("Valeurs manquantes :")
    print(df.isna().sum().head(10))  # juste les 10 premières colonnes
    print("Doublons :", df.duplicated().sum())
    print("Aperçu des 5 premières lignes :")
    print(df.head(5))
```
